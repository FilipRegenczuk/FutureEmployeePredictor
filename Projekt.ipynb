```python
# Logging

from getpass import getpass

def logging():
    email = input("Email:")
    password = getpass("Password:")
    user = None
    
    if not email:
        print("\nEnter email!")
    else:
        data = cur.execute("SELECT * FROM Users WHERE Email=?", (email,)).fetchone()
        if not (data):
            print("\nUser with email not found!")
        else:
            if (data[-1] != password):
                print("\nWrong password!")
            else:
                user = data
                print(f"\nWelcome {user[1]}!")
    
    return list(user)
    
```

```python
from keras.models import load_model

model = load_model('model')
```

```python
model.summary()
```

```python
def prepare_array_for_model(user):
    
    # mapping dictionaries
    education = {
        "Bachelor": 1,
        "Masters": 2,
        "PHD": 3
    }
    city = {
        "Bangalore": 1,
        "New Delhi": 2,
        "Pune": 3
    }
    gender = {
        "Male": 0,
        "Female": 1
    }
       
    # Process list
    # 1. Trim unnecessary params
    array = user[3:-2]
    # 2. Transform string to mappers
    array[0] = [education[i] for i in array if i in education.keys()][0]
    array[1] = [city[i] for i in array if i in city.keys()][0]
    array[3] = [gender[i] for i in array if i in gender.keys()][0]
    
    return np.array(array).reshape([1,len(array)])
```

```python
# Predict if employee will leave firm in 2 years

def predict_leaving(user):
    array = prepare_array_for_model(user)
    y_pred = (model.predict(array) > 0.5).astype(int)
    return bool(y_pred.tolist()[0][0])
```

```python
print("Predicting if you will leave firm in 2 years:")
predict_leaving(user)
```
